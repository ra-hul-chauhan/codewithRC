# CLEANING

After analysing we found oout that our data is not perfect ,some columns are irrelevnt , some has missing data and some has null values , so we will clean our data or convert it into usable one or format by pandas to perform further operations

```python
# rename column
a = weather.rename(columns={'WindSpeed' : 'windspeedAVG'})
a.head(10)
```

```python
#filling missing values 
weather= weather.fillna(weather.mean())
weather
```

```python
#dropping unnecessary columns 
weather=weather.drop(columns=["Solare Radiation AVG"])
weather.head(10)
```

```python
#find correlation matrix
x = weather[['Date','Battery Voltage AVG','Temperature AVG','Relative Humidity AVG','Wind Direction AVG']].corr()
x
#negative value depicts that columns are inversely proportional and vica versa
#in correlation range is from -1 to +1 always
```

# DATA VISUALIZATION 

```python
import matplotlib.pyplot as plt
%matplotlib inline
#
y1 = weather['Relative Humidity AVG'].head(50)
x = range(50)
plt.plot(x,y1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#
y1 = weather['Relative Humidity AVG'].head(50)
y2 = weather['Temperature AVG'].head(50)
x = range(50)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend()
```

```python
y1 = weather['Relative Humidity AVG'].head(50)
y2 = weather['Temperature AVG'].head(50)
x = range(50)
plt.stackplot(x,y1,alpha = .5,color = 'orange')
plt.stackplot(x,y2,alpha = .7,color = 'green')
```

```python
# plotting with both lineplot and stackplot(area plot) to see the margin
y1 = weather['Relative Humidity AVG'].head(50)
y2 = weather['Temperature AVG'].head(50)
x = range(50)
plt.plot(x,y1,linewidth = 2.0, color = 'c')
plt.stackplot(x,y1,alpha = .5,color = 'orange')
plt.plot(x,y2,linewidth = 1.0, color = 'r')
plt.stackplot(x,y2,alpha = 1,color = 'green')
```
